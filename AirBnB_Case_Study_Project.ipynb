```python
import numpy as np
import pandas as pd

air_b = pd.read_csv("/Users/kali/Documents/Project_3/AB_NYC_2019.csv")
```

```python
display(air_b)
```

```python
# How many neighborhood groups are available and which shows up the most?

air_b.groupby(["neighbourhood_group"]).count()["id"].to_frame()
```

```python
# Are private rooms the most popular in manhattan?
air_b.groupby(["room_type", "neighbourhood_group"]).count()["id"].to_frame()

```

```python
# Which hosts are the busiest and based on their reviews?

air_b.groupby(["host_name", "number_of_reviews"]).count()["id"].sort_values(ascending=False).head(10)
```

```python
#Which neighorhood group has the highest average price?

air_b.groupby(["neighbourhood_group"])["price"].mean().sort_values(ascending=False).round().head(10
                                                                                                 )
```

```python
# Which neighbor hood group has the highest total price?

air_b.groupby(["neighbourhood_group"])["price"].max().sort_values(ascending=False)
```

```python
#Which top 5 hosts have the highest total price?

air_b.groupby(["host_name"])["price"].max().sort_values(ascending=False).head(10)

```

```python
# Who currently has no (zero) availability with a review count of 100 or more?

#air_b.groupby("host_name").count()[["availability_365", "number_of_reviews"]].sort_values("availability_365", ascending=True)

air_b[(air_b["number_of_reviews"] >=100) & (air_b["availability_365"] == 0)].groupby(["host_name"], as_index=False).count().sort_values("number_of_reviews", ascending=False)
```

```python
# What host has the highest total of prices and where are they located?

air_b.groupby(["host_name", "neighbourhood_group"])["price"].max().sort_values(ascending=False).head(10)

```

```python
# When did Danielle from Queens last receive a review?

air_b["last_review"] = pd.to_datetime(air_b["last_review"])

air_b[(air_b["host_name"] == "Danielle")].groupby(["host_name"])["last_review"].max()



```
